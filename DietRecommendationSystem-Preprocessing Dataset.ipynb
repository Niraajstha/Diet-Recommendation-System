"""Removes the c("...") format and extra characters."""

```python
def clean_c_format(value):
    if pd.isna(value):
        return None
    return re.sub(r'[c\(\)"\']', '', value).strip()

dataset['Keywords'] = dataset['Keywords'].apply(clean_c_format)
dataset['RecipeIngredientParts'] = dataset['RecipeIngredientParts'].apply(clean_c_format)
dataset['RecipeIngredientQuantities'] = dataset['RecipeIngredientQuantities'].apply(clean_c_format)
```

"""Converts c("...") formatted instructions into a list of steps."""

```python

def clean_instructions(value):
    if pd.isna(value):
        return None
    steps = re.findall(r'"(.*?)"', value)
    return " ".join(steps) if steps else value

dataset['RecipeInstructions'] = dataset['RecipeInstructions'].apply(clean_instructions)
```
